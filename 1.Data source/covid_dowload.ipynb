```python
import os
import requests
import json

def obtener_lista_codigos_estado():
    try:
        url = "https://api.covidtracking.com/v1/states/info.json"
        response = requests.get(url)
        response.raise_for_status()  # Verificar si la solicitud fue exitosa
        lista_estados = response.json()
        codigos_estado = [estado['state'] for estado in lista_estados]
        return codigos_estado
    except requests.exceptions.RequestException as e:
        print("Error al obtener la lista de códigos de estado:", e)
        return []

def obtener_registro_estado(codigo_estado):
    try:
        url = f"https://api.covidtracking.com/v1/states/{codigo_estado.lower()}/current.json"
        response = requests.get(url)
        response.raise_for_status()  # Verificar si la solicitud fue exitosa
        registro_estado = response.json()
        return registro_estado
    except requests.exceptions.RequestException as e:
        print(f"Error al obtener el registro del estado {codigo_estado}:", e)
        return None

def guardar_en_json(datos, nombre_archivo):
    try:
        # Obtener la ruta completa del directorio 'json'
        carpeta_json = './json'
        # Verificar si la carpeta 'json' existe, si no, crearla
        if not os.path.exists(carpeta_json):
            os.makedirs(carpeta_json)
        # Unir la ruta de la carpeta con el nombre del archivo
        ruta_completa = os.path.join(carpeta_json, nombre_archivo)
        # Guardar los datos en el archivo JSON
        with open(ruta_completa, 'w') as archivo:
            json.dump(datos, archivo, indent=4)
        print("Datos guardados exitosamente en", ruta_completa)
    except IOError as e:
        print("Error al guardar el archivo:", e)

# Obtener la lista de códigos de estado
codigos_estados = obtener_lista_codigos_estado()

# Lista para almacenar todos los registros
todos_los_registros = []

for codigo_estado in codigos_estados:
    # Obtener el registro del estado específico
    registro_estado = obtener_registro_estado(codigo_estado)
    if registro_estado:
        todos_los_registros.append(registro_estado)
    else:
        print(f"No se pudieron obtener los datos para el estado {codigo_estado}.")

# Guardar todos los registros en un solo archivo JSON
guardar_en_json(todos_los_registros, "registroscovid.json")
```

```python
import os
import requests
import json
from datetime import datetime, timedelta

def obtener_lista_codigos_estado():
    try:
        url = "https://api.covidtracking.com/v1/states/info.json"
        response = requests.get(url)
        response.raise_for_status()  # Verificar si la solicitud fue exitosa
        lista_estados = response.json()
        codigos_estado = [estado['state'] for estado in lista_estados]
        return codigos_estado
    except requests.exceptions.RequestException as e:
        print("Error al obtener la lista de códigos de estado:", e)
        return []

def obtener_registro_estado(codigo_estado, fecha):
    try:
        url = f"https://api.covidtracking.com/v1/states/{codigo_estado.lower()}/{fecha}.json"
        response = requests.get(url)
        response.raise_for_status()  # Verificar si la solicitud fue exitosa
        registro_estado = response.json()
        return registro_estado
    except requests.exceptions.RequestException as e:
        print(f"Error al obtener el registro del estado {codigo_estado} para la fecha {fecha}:", e)
        return None

def guardar_en_json(datos, nombre_archivo):
    try:
        # Obtener la ruta completa del directorio 'json'
        carpeta_json = './json'
        # Verificar si la carpeta 'json' existe, si no, crearla
        if not os.path.exists(carpeta_json):
            os.makedirs(carpeta_json)
        # Unir la ruta de la carpeta con el nombre del archivo
        ruta_completa = os.path.join(carpeta_json, nombre_archivo)
        # Guardar los datos en el archivo JSON
        with open(ruta_completa, 'w') as archivo:
            json.dump(datos, archivo, indent=4)
        print("Datos guardados exitosamente en", ruta_completa)
    except IOError as e:
        print("Error al guardar el archivo:", e)

# Obtener la lista de códigos de estado
codigos_estados = obtener_lista_codigos_estado()

# Definir la fecha de inicio (4 meses antes del 7 de marzo de 2021)
fecha_inicio = datetime(2021, 3, 1)

# Lista para almacenar todos los registros
todos_los_registros = []

# Iterar sobre los últimos cuatro meses
fecha_actual = datetime(2021, 3, 7)
while fecha_actual >= fecha_inicio:
    fecha_actual_str = fecha_actual.strftime("%Y%m%d")
    for codigo_estado in codigos_estados:
        # Obtener el registro del estado específico para la fecha dada
        registro_estado = obtener_registro_estado(codigo_estado, fecha_actual_str)
        if registro_estado:
            todos_los_registros.append(registro_estado)
        else:
            print(f"No se pudieron obtener los datos para el estado {codigo_estado} en la fecha {fecha_actual_str}.")
    fecha_actual -= timedelta(days=1)

# Guardar todos los registros en un solo archivo JSON
guardar_en_json(todos_los_registros, "registroscovid.json")
```
