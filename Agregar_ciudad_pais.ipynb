```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit, col, rand, when
import json

# Iniciar sesión de Spark
spark = SparkSession.builder \
    .appName("Agregar ciudad y país a registros COVID") \
    .getOrCreate()

# Leer el archivo de ciudades JSON
df_ciudades = spark.read.json("./1.Data source/json/ciudades.json", mode="PERMISSIVE", multiLine="true")

# Leer el archivo de países JSON
df_paises = spark.read.json("./1.Data source/json/paises.json", mode="PERMISSIVE", multiLine="true")

# Leer el archivo "registroscovid.json"
df_registroscovid = spark.read.json("./1.Data source/json/registroscovid.json", mode="PERMISSIVE", multiLine="true")

# Registrar DataFrames como tablas temporales
df_ciudades.createOrReplaceTempView("ciudades")
df_paises.createOrReplaceTempView("paises")
df_registroscovid.createOrReplaceTempView("registroscovid")

# Obtener una lista de nombres de ciudades únicas
nombres_ciudades = df_ciudades.select("nombre").distinct().rdd.flatMap(lambda x: x).collect()

# Lista para almacenar todos los DataFrames con la columna ciudad y país agregada
dfs_con_ciudad_pais = []

# Iterar sobre cada nombre de ciudad y aplicar la lógica
for nombre_ciudad in nombres_ciudades:
    df_registro_con_ciudad = df_registroscovid.withColumn("ciudad", lit(nombre_ciudad))
    nombre_pais = spark.sql(f"SELECT nombre FROM paises WHERE id_pais = (SELECT id_pais FROM ciudades WHERE nombre = '{nombre_ciudad}')").collect()[0][0]
    df_registro_con_ciudad_pais = df_registro_con_ciudad.withColumn("pais", lit(nombre_pais))
    dfs_con_ciudad_pais.append(df_registro_con_ciudad_pais)

# Unir todos los DataFrames en uno solo
df_final = dfs_con_ciudad_pais[0]
for df in dfs_con_ciudad_pais[1:]:
    df_final = df_final.union(df)

# Definir el rango del factor multiplicador
min_factor = 1  # Cambia esto por el valor mínimo que desees
max_factor = 10  # Cambia esto por el valor máximo que desees

# Aplicar el factor multiplicador aleatorio a cada columna numérica, excepto la columna 'date'
for col_name in df_final.columns:
    if col_name != 'date' and col_name != 'ciudad' and col_name != 'pais':
        df_final = df_final.withColumn(col_name, when(col("date").isNull(), col(col_name)).otherwise((col(col_name) * (min_factor + (rand() * (max_factor - min_factor)))).cast('bigint')))

# Convertir el DataFrame de Spark a un DataFrame de pandas
df_pandas = df_final.toPandas()

# Guardar el DataFrame de pandas en un archivo JSON con orientación de registros
df_pandas.to_json('./2.Json+ciudad+pais/df_con_ciudad_pais_registroscovid.json', orient='records')

# Mostrar el DataFrame final
df_final.show()
```
