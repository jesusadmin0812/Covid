```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, year, month, dayofmonth, to_date, lit, monotonically_increasing_id
from datetime import datetime

# Iniciar sesión de Spark
spark = SparkSession.builder \
    .appName("Guardar datos en Parquet histórico") \
    .getOrCreate()

# Leer el archivo JSON en un DataFrame Spark
df_final = spark.read.json("./2.Json+ciudad+pais/df_con_ciudad_pais_registroscovid.json")

# Generar la fecha actual
fecha_procesado = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Convertir la columna 'date' al formato de fecha adecuado
df_final = df_final.withColumn("date", to_date(col("date").cast("string"), "yyyyMMdd"))

# Agregar la fecha de procesamiento como una nueva columna
df_final = df_final.withColumn("fecha_procesado", lit(fecha_procesado))

# Extraer año, mes y día de la fecha y agregarlos como columnas separadas
df_final = df_final.withColumn("aaaa", year("date"))
df_final = df_final.withColumn("mmm", month("date"))
df_final = df_final.withColumn("dd", dayofmonth("date"))

# Agregar una columna index que vaya sumando de uno en uno
df_final = df_final.withColumn("id", monotonically_increasing_id())

# Agregar la columna "parametro" con el valor constante 1
df_final = df_final.withColumn("parametro", lit(1))

# Particionar los datos por país y ciudad
df_final.write.partitionBy("pais", "ciudad").parquet("./3.parquet_historico", mode="overwrite")

# Mostrar el DataFrame final
df_final.show()
```
